```python
# Read in the vine_table.csv file as a DataFrame, and perform your analysis using the steps above.
vine_data = r"C:\Users\mhern\Desktop\Analysis_Projects_2021\Module_16/vine_table.csv"
vine_df = pd.read_csv(vine_data)
vine_df
```

```python
#1.Filter the data and create a new DataFrame or table to retrieve all the rows where the total_votes count is equal to or greater than 20 to pick reviews that are more likely to be helpful and to avoid having division by zero errors later on.

total_votes_20 = vine_df[(vine_df["total_votes"]>=20)]
total_votes_20
```

```python
#2.Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
total_votes_20["helpful_votes_percent"] = total_votes_20["helpful_votes"]/total_votes_20["total_votes"]
total_votes_20
```

```python
vine_program = total_votes_20[(total_votes_20)["helpful_votes_percent"]>=0.5]
vine_program
```

```python
#3.Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves all the rows where a review was written as part of the Vine program (paid), vine == 'Y'.
vine_program_paid = vine_program[(vine_program["vine"]== 'Y')]
vine_program_paid["review_id"].count()
```

```python
#4.Repeat Step 3, but this time retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'.
vine_program_unpaid = vine_program[(vine_program["vine"]== 'N')]
vine_program_unpaid["review_id"].count()
```

```python
#5.Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).
#numer of reviews 
vine_program["review_id"].count()
```

```python
#the number of 5-star reviews paid 
vine_five_paid = vine_program_paid[vine_program_paid["star_rating"] == 5]
vine_five_paid["review_id"].count()
```

```python
#the number of 5-star reviews unpaid
vine_five_unpaid = vine_program_unpaid[vine_program_unpaid["star_rating"] == 5]
vine_five_unpaid["review_id"].count()
```

```python
#the number of 5-star reviews paid percent
vine_five_paid_percent = (vine_five_paid["review_id"].count()/vine_program_paid["review_id"].count())*100
vine_five_paid_percent
```

```python
#the number of 5-star reviews paid percent
vine_five_unpaid_percent = (vine_five_unpaid["review_id"].count()/vine_program_unpaid["review_id"].count())*100
vine_five_unpaid_percent
```
